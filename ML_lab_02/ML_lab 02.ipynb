```python
import warnings
warnings.filterwarnings('ignore')
import tensorflow as tf
```

```python
# train data set
x_train = [1, 2, 3]
y_train = [1, 2, 3]

W = tf.Variable(tf.random_normal([1]), name = 'weight') # tensor의 shape을 준다.
b = tf.Variable(tf.random_normal([1]), name = 'bias')

# Our hypothesis XW + b
hypothesis = x_train * W + b
```

```python
 # cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - y_train))
# _ reduce_mean은 그냥 평균을 내는 것
```

# GradientDescent

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer()) # variable을 사용하기 위해서
```

```python
for step in range(1001):
    sess.run(train)
    if step % 20 == 0:
        print(step,":" ,sess.run(cost), sess.run(W), sess.run(b))
```

# Placeholders

```python
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

for step in range(2001):
    cost_val, W_val, b_val, _ = sess.run([cost, W, b, train], feed_dict = {X: [1, 2, 3, 4, 5], Y: [1, 2, 3, 4, 5]})
    if step % 20 == 0:
        print(step, cost_val, W_val, b_val)
```
